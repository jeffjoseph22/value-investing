# Exploratory Analysis 

## Create a heat map 
to get a general sense of the relationships between the variables 

```python
# Get Correlation Values 
pearson_corr = df.corr()

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))
plt.title("Heat Map Comparing the Pearson Correlations of All Features")
# Generate a mask hide the upper triangle
mask = np.triu(np.ones_like(pearson_corr, dtype=bool))

# Generate a custom diverging colormap
cmap = sb.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sb.heatmap(pearson_corr, cmap=cmap, mask = mask, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
# Look at correlation with marketCap (ideal predictor variable)
pearson_corr['marketCap']
```

### Spearman Heat Map

```python

# Get Correlation Values 
spearman_corr = df.corr(method='spearman')

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))
plt.title("Heat Map Comparing the Spearman Correlations of All Features")
# Generate a mask hide the upper triangle
mask = np.triu(np.ones_like(spearman_corr, dtype=bool))

# Generate a custom diverging colormap
cmap = sb.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sb.heatmap(spearman_corr, cmap=cmap, mask = mask, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
# Look at correlation with marketCap (ideal predictor variable)
print("=====\t Pearson Correlation w/ Market Cap\t=====")
pearson_corr['marketCap']
print("=====\t Spearman Correlation w/ Market Cap\t=====")
spearman_corr['marketCap']
```

# Machine Learning

```python
# Plus linear regression, decision tree, support vector regression, lasso regression, random forest, polynomial, stepwise, multiple linear

# import split function
from sklearn.model_selection import train_test_split

# import models 
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn import svm
from sklearn.linear_model import Lasso
from sklearn.ensemble import RandomForestRegressor

# import metrics 
from sklearn.metrics import mean_squared_error as MSE
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score


# Slim down df to just vars of current interest
ml_df = df[['netIncome', 'netDebt', 'freeCashFlow', 'currentDebt', 'marketCap']]
ml_df = ml_df.drop(['currentDebt'], axis=1) # get rid of currentDebt because it has a lot of nan's and they don't fly with the machine learning
# now drop the nans 
ml_df = ml_df.dropna()

# Create the predictor variables 
X = ml_df.drop(['marketCap'], axis = 1)

# The independent variable 
y = ml_df['marketCap']

# Create training and testing data 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 3)

y_train.shape
```

```python
# Create the models
LR_model = LinearRegression()
DT_model = DecisionTreeRegressor()
# SVM_model = svm.svm
LASSO_model = Lasso(alpha=0.1)
RFR_model = RandomForestRegressor(n_estimators = 50)


# Create a list of the models
model_list = [LR_model, DT_model, LASSO_model, RFR_model]

# run each model 
for model in model_list:
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    print(str(model))
    print("\t test RMSE:", round(MSE(y_test, predictions)**0.5, 2))
    print("\t test corr:", np.corrcoef(y_test, predictions)[0,1])
    print("\t r2:", r2_score(y_test, predictions), "\n")
```

# Tenative Results 
The best two models are Decision tree and random forrest regression. 
- The Decision tree had an r^2 of .9 (the closer to 1 or -1 the better, so a great score).
- Random Forest had an r^2 of .88 (also a good score )

I'm going to repeat the test for random forest with more estimators to see if that helps. Then I'll look at more metrics of the models performance and then try to visualize the results. 

```python
estimators = [50, 100, 150, 200, 250]
# estimators = [150]
for i in estimators:
    model = RandomForestRegressor(n_estimators = i).fit(X_train, y_train)
    predictions = model.predict(X_test)

    print("Number of Estimators:", i)
    print("\t RMSE: ", MSE(y_test, predictions)**0.5)
    print("\t Correlation: ", np.corrcoef(y_test, predictions)[0,1])
    print("\t R-squared: ", r2_score(y_test, predictions))
```

# Random Forsest Results
With 100 estimators, we got the best r-squared of .90  (still not better than the decision tree)

```python
# import more metrics 
from sklearn.metrics import mean_absolute_error as MAE
from sklearn.metrics import mean_absolute_percentage_error as MAPE

RFR = RandomForestRegressor(n_estimators=100)
DTR = DecisionTreeRegressor()

models_of_interest = [RFR, DTR]
for model in models_of_interest:
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    print(str(model))
    print("\t test RMSE:", round(MSE(y_test, predictions)**0.5, 2))
    print("\t test corr:", np.corrcoef(y_test, predictions)[0,1])
    print("\t r2:", r2_score(y_test, predictions), "\n")
    print("\t Mean Absolute Error:", MAE(y_test, predictions))
    print("\t Mean Absolute Percent Error:", MAPE(y_test, predictions))
    print("\t Mean Error:", np.mean(predictions - y_test), '\n\n')
```
