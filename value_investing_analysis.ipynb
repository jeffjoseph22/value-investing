# Import Data

```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv('company-data.csv')
data
```

# Clean 

```python
data = data.rename(columns={"Unnamed: 0":"date"})
data
```

# Exploratory Analysis 

## Create a heat map 
to get a general sense of the relationships between the variables 

```python


# Get Correlation Values 
corr = data.corr()

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a mask hide the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Generate a custom diverging colormap
cmap = sb.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sb.heatmap(corr, cmap=cmap, mask = mask, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
# Look at correlation with marketCap (ideal predictor variable)
corr['marketCap']
```

## Visualize Market Cap

```python
latest_q_data = data.query("date == '2022-03-31'", )
```

```python
plt.figure(dpi=350)
plt.bar(latest_q_data['Company'], height=latest_q_data['marketCap'])
plt.xlabel('Coompany')
plt.ylabel("Market Cap (In Trillions)")
plt.title("Market Cap of Companies")
plt.show()
```
